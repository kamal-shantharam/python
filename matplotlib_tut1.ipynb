# Matplot library

````python
from matplotlib import pylab
print(pylab.__version__)
````

````python
# Using numpy to generage random numbers
import numpy as np
x = np.linspace(0, 10, 25)
y = x * x + 2
print(x)
print(y)
print(np.array([x, y]).reshape(25,2))
print(np.array([x, y]).reshape(25,2).reshape(2, 25))
# draw plot
pylab.plot(x, y, 'r')
# subgraph
pylab.subplot(1, 2, 1)
pylab.plot(x, y, 'r--')
pylab.subplot(1, 2, 2)
pylab.plot(x, y, 'g*-')
````

````python
# using matplotlib pyplot

from matplotlib import pyplot as plot

fig = plot.figure(figsize=(16, 9), dpi=600)
axes = fig.add_axes([0.5, 0.1, 0.8, 0.8])
axes.plot(x, y, 'r')
fig, axes = plot.subplots(nrows=1, ncols=2)
for ax in axes:
  ax.plot(x, y, 'r')

# draw a picture/graph

fig = plot.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
````
